```python
#Exercise 1: Pizza Toppings 

#Make a program that ask the user, on what toppings they want to add on their pizza
topping_name='\nPlease enter the name of the topping you would like to add:'
topping_name+='\nEnter quit when you are finished'

#Using a while-if-else loop and use break to stop the loop in this program
while True:
  pizza=input(topping_name)

  if pizza=='quit':
     break
  else:
    print('I would love to add', pizza.title(),'!')
```

```python
#Exercise 4: Deli 

#Create a sandwich orders and fill it with the names of various number
sandwich_orders = ['Chicken', 'Tuna', 'Egg', 'Ham']
finished_sandwiches = []

#Use the while loop for sandwich_orders and print a sentence that you made the user's sandwich
while sandwich_orders:
  current_sandwich = sandwich_orders.pop()
  print("I made your " + current_sandwich + " sandwich.")
  finished_sandwiches.append(current_sandwich)

print("\n")
print('Sandwich Orders:')

#Use the for loop for finished_sandwiches
for sandwich in finished_sandwiches:
  print(sandwich,'Sandwich')
```

```python
#Exercise 5: No Pastrami

#Make a program that shows that deli has run out of Pastrami
print('The deli has run out of Pastrami.')
print('\n')

#Make a sandwich orders using sandwich_orders and select a various numbers
sandwich_orders = ['Chicken', 'Tuna', 'Egg', 'Ham']
finished_sandwiches = []

#Use the for-if loop to remove the pastrami on the list
for sandwich in sandwich_orders:
  if sandwich == 'Pastrami':
    sandwich_orders.remove(sandwich)

#Use while loop to show to the users their made sandwiches
while sandwich_orders:
  current_sandwich = sandwich_orders.pop()
  print("I made your " + current_sandwich + " sandwich.")
  finished_sandwiches.append(current_sandwich)

print("\n")
print('Sandwich Orders:')

#Use for loop for finished_sandwiches
for sandwich in finished_sandwiches:
  print(sandwich,'Sandwich')

for sandwich in finished_sandwiches:
  print(sandwich,'Pastrami')
```
